# Different Graphical Styles with MatPlotLib

See details here: https://matplotlib.org/

```python
import pandas as pd  # A data analysis library
%matplotlib inline
import matplotlib.pyplot as plt  # Python's plotting library
import seaborn as sns  # Makes matplotlib prettier
import numpy as np
your_file = "hypothetical.csv" # Make sure that your desired csv is in the same folder as this notebook.
csv_hypo = pd.read_csv(your_file)  # delete the first hash tag of this line!
csv_hypo = csv_hypo.set_index('Date') #Set this string as the name of your dates column
csv_hypo  # delete the first hastag of this line!
```

# Markers and Color Maps

To place markers on your lines, use the marker argument. Documentation is here: 
https://matplotlib.org/examples/lines_bars_and_markers/marker_reference.html

Let's find some more compelling colormaps! There's a great reference here:
https://matplotlib.org/examples/color/colormaps_reference.html

```python
mygraph = csv_hypo.plot(kind='line', rot=30, title='Sleep vs Walk', marker='o', colormap='Set2');
mygraph
```

```python
myDualAxisgraph = csv_hypo['Sleep'].plot(kind='line', rot=30, title='Sleep vs Drive', marker='o', fillstyle = 'full', color='#aaaaaa', linewidth=3);
myDualAxisgraph.set_ylabel('Sleep (hours)', color='#aaaaaa');
myDualAxisgraph.spines['top'].set_visible(False)
myDualAxisgraph.tick_params('y', colors='#aaaaaa');

myDualAxisgraph2 = myDualAxisgraph.twinx();

myDualAxisgraph2 = csv_hypo['Drive'].plot(kind='line', marker='o', fillstyle = 'full', color='b', linewidth=1);
myDualAxisgraph2.spines['top'].set_visible(False)
myDualAxisgraph2.set_ylabel('Drive (miles)', color='b');
myDualAxisgraph2.tick_params('y', colors='b');
```
